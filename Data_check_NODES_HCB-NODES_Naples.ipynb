# Saving and loading DFs as pkl files

```python
def save_dataframe(df, filename, file_format='pkl'):
    """
    Save a pandas DataFrame in the desired format.

    Parameters:
    - df: pandas DataFrame
    - filename: name of the output file (without extension)
    - file_format: 'pkl', 'csv', 'parquet', or 'feather' (default: 'pkl')
    """
    if file_format == 'pkl':
        df.to_pickle(f"{filename}.pkl")
        print(f"âœ… DataFrame saved as {filename}.pkl")
    elif file_format == 'csv':
        df.to_csv(f"{filename}.csv", index=False)
        print(f"âœ… DataFrame saved as {filename}.csv")
    elif file_format == 'parquet':
        df.to_parquet(f"{filename}.parquet")
        print(f"âœ… DataFrame saved as {filename}.parquet")
    elif file_format == 'feather':
        df.to_feather(f"{filename}.feather")
        print(f"âœ… DataFrame saved as {filename}.feather")
    else:
        print("âŒ Unsupported file format. Choose 'pkl', 'csv', 'parquet', or 'feather'.")

# Example usage:
# save_dataframe(clinic_hcb_df, 'clinic_data', file_format='parquet')

from pathlib import Path

ruta_naples = Path('F:/Cursos/UOC Master Bio Inf. Est/M0.209 - UOC - TFM - Bio ML - Grafos/DATOS/MULTILAYER/DADES_NAPLES') 
ruta_hcb = Path('F:/Cursos/UOC Master Bio Inf. Est/M0.209 - UOC - TFM - Bio ML - Grafos/DATOS/MULTILAYER/DADES_HCB') 

# Save
save_dataframe(nodes_naples_df, ruta_naples / 'NODES_NAPLES', file_format='pkl')
save_dataframe(nodes_hcb_df, ruta_hcb / 'NODES_HCB', file_format='pkl')

```

```python
def load_dataframe(filepath):
    """
    Load a pandas DataFrame from a file, inferring the format by extension.

    Parameters:
    - filepath: full path to the file (with extension)

    Returns:
    - pandas DataFrame
    """
    if str(filepath).endswith('.pkl'):
        df = pd.read_pickle(filepath)
    elif str(filepath).endswith('.csv'):
        df = pd.read_csv(filepath)
    elif str(filepath).endswith('.parquet'):
        df = pd.read_parquet(filepath)
    elif str(filepath).endswith('.feather'):
        df = pd.read_feather(filepath)
    else:
        raise ValueError("âŒ Unsupported file format. Supported formats: .pkl, .csv, .parquet, .feather")
    
    print(f"âœ… DataFrame loaded from {filepath}")
    return df
# Example usage:
# df_loaded = load_dataframe(rute / 'clinic_data.parquet')


# Load the saved DataFrames
df_loaded_NODES_NAPLES = load_dataframe(ruta_naples / 'NODES_NAPLES.pkl')
df_loaded_NODES_HCB = load_dataframe(ruta_hcb / 'NODES_HCB.pkl')
```
